# count plot in seaborn

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.countplot(x="sex",data=var)
plt.show()
#yaha per counter plot me hame counting value dikhai de rahi hai.
```

```python
sns.barplot(x="sex",y="size",data=var)
plt.show()
# or bar plot me hame mean value dikhai deti hai
```

```python
sns.countplot(x="sex",data=var,hue="smoker")
plt.show()
```

```python
# Agar hame graph ko verticle se horizontal banana hai to vo bhi hum yaha per kar sakte hai.
# uske liye hame yaha per x parameter ki jagah per y parameter ko dena hoga
sns.countplot(y="sex",data=var,hue="smoker")
plt.show()
```

```python
#palette parameter
sns.countplot(x="sex",data=var,hue="smoker",palette="bwr")
plt.show()
```

```python
#color parameter
sns.countplot(x="sex",data=var,hue="smoker",color="r")
plt.show()
```

```python
# agar hame graph ko thoda sa light dikhana hai to uske liye hum saturation parameter ka use kar sakte hai.
# saturation parameter
sns.countplot(x="sex",data=var,hue="smoker",color="r",saturation=0.5)
plt.show()
```
