# strip plot in seaborn

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
var=sns.load_dataset("tips")
var
```

```python
sns.stripplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
# differentiating according to the male and female
sns.stripplot(x="day",y="total_bill",data=var,hue="sex")
plt.show()
```

```python
# palette parameter 
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette='rocket_r')
plt.show()
```

```python
# changing the edgecolor and linewidth or size
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",linewidth=1.5,edgecolor="r",size=9)
plt.show()
```

```python
# hum yaha per edge ke beech me jo displacement hai usko, hum kam ya jyada kar sakte hai.
# iske liye hum jitter parameter ka use karte hai.
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",linewidth=1.5,jitter=5)
plt.show()
```

```python
# changing the marker
# m={"Male":"*","Female":"o"}
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",marker="<")
plt.show()
```

```python
# agar hame hamare graph ko thoda sa light dikhana hai to uske liye hum yaha per alpha parameter ka use kar sakte hai.
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",marker="<",alpha=0.7)
plt.show()
```

```python
# agar hum chahte hai ki yaha per hame sirf ek hi column ka distint plot mile to hum vo bhi kar sakte hai.
sns.stripplot(x=var["total_bill"])
plt.show()
```

```python
#for verticle graph
sns.stripplot(y=var["total_bill"])
plt.show()
```
