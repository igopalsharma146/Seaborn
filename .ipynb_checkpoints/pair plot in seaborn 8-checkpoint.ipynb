# pair plot in seaborn

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
var=sns.load_dataset("tips")
var
```

```python
sns.pairplot(var)
plt.show()
```

```python
sns.pairplot(var,hue="sex")
plt.show()
```

```python
# if we are want to make pair plot of few columns
sns.pairplot(var,vars=["total_bill","tip"],hue="sex")
plt.show()
```

```python
# hum yaha per iske order ko bhi set kar sakte hai.
sns.pairplot(var,vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"])
plt.show()
```

```python
# hum yaha per iske color ko bhi change kar sakte hai . uske liye hum palette paramaeter ka use kar sakte hai
sns.pairplot(var,vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"],palette="Dark2")
plt.show()
```

```python
# hum yaha per graph ke variable ko bhi change kar saakte hai. ki x axis ke ander konsa hoga or y axis ke ander konsa hoga.

sns.pairplot(var,hue="sex",hue_order=["Female","Male"],x_vars=["total_bill","tip"])
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],y_vars=["total_bill","tip"])
plt.show()
```

```python
# hum yaha per graph ko bhi change kar sakte hai. uske liye ham kind parameter ka use karte hai.
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],kind="reg") # 'scatter','kde','hist','reg'
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],kind="kde")
plt.show()
```

```python
# diagonal plot ko bhi hum yaha per fix kar sakte hai.
# uske liye hame diag_kind parameter ka use karna hoga.
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],kind="reg",diag_kind="hist")
plt.show()
```

```python
# changing the marker
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],kind="reg",markers=["*","<"])
plt.show()
```
