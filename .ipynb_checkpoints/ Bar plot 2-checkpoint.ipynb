# Bar plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var=sns.load_dataset("penguins")
var
```

```python
sns.barplot(x=var.island,y=var.bill_length_mm)
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=var)
plt.show()
```

```python
# if we want the make different graph of male and female, then we can use here hue parameter
# hue parameter ki help se hum hum graph ko differentiate kar sakte hai.
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex")
plt.show()
```

```python
#changing the order
order1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",order=order1)
plt.show()
```

```python
#changing the order of male and female
order1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",order=order1,hue_order=["Female","Male"])
plt.show()
```

```python
# ci parameter
order1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",order=order1,
            hue_order=["Female","Male"],ci=20)
plt.show()
```

```python
# n_boot parameter
order1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",order=order1,
            hue_order=["Female","Male"],n_boot=2)
plt.show()
```

```python
# orientation
# NOTE:if we are printing the horizontal graph then hum jo bhi variable dende vo dono variable numericle type ke hone chahiye
order1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="bill_depth_mm",y="bill_length_mm",data=var,orient="v")
plt.show()
```

```python
order1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="bill_depth_mm",y="bill_length_mm",data=var,orient="h")
plt.show()
```

```python
order1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,color="r")
plt.show()
```

```python
order1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,color="r",palette='icefire')
plt.show()
```

```python
#saturation parameter
order1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",order=order1,
            hue_order=["Female","Male"],saturation=.5)
plt.show()
```

```python
#saturation parameter
order1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",order=order1,
            hue_order=["Female","Male"],saturation=10)
plt.show()
```

```python
#changing the color and width of error
order1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",order=order1,
            hue_order=["Female","Male"],saturation=5,errcolor="g",errwidth=10)
plt.show()
```

```python
#incresing the capsize
order1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",order=order1,
            hue_order=["Female","Male"],saturation=5,errcolor="g",capsize=.5)
plt.show()
```

```python
#dodge parameter
order1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",order=order1,
            hue_order=["Female","Male"],saturation=5,errcolor="g",dodge=False)
plt.show()
```

```python
#alpha parameter
order1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",order=order1,
            hue_order=["Female","Male"],saturation=5,errcolor="g",alpha=0.8)
plt.show()
```

```python
#sns grid , hum yaha per plt.grid() ka bhi use kar sakte hai
order1=["Dream","Torgersen","Biscoe"]
sns.set(style="dark")

sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",order=order1,
            hue_order=["Female","Male"],saturation=5,errcolor="g",alpha=0.8)
plt.show()
```

```python
#sns grid , hum yaha per plt.grid() ka bhi use kar sakte hai
order1=["Dream","Torgersen","Biscoe"]
sns.set(style="darkgrid")

sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",order=order1,
            hue_order=["Female","Male"],saturation=5,errcolor="g",alpha=0.8)
plt.show()
```
