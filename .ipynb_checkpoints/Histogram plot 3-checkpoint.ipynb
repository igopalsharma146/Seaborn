# Histogram plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var=sns.load_dataset("penguins")
var
```

```python
sns.displot(var["bill_length_mm"])
plt.show()
```

```python
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240])
plt.show()
```

```python
# kde=kernal density estimator
#kde ki help se hum bar graph ke uper ek line graph show kara sakte hai.
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240],kde=True)
plt.show()
```

```python
# rug parameter
#rug parameter ki help se hum kernal density ko x-axis per show kara sakte hai.
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240],kde=True,rug=True)
plt.show()
```

```python
#changing the color
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240],kde=True,rug=True,color="r")
plt.show()
```

```python
# log_scale
# agar hame data ko log ke ander convert karna hai to hum uske liye log_scale ka use kar sakte hai.
sns.displot(var["body_mass_g"],bins=[170,180,190,200,210,220,230,240],kde=True,rug=True,color="r",log_scale=True)
plt.show()
```
