# Scatter plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var=sns.load_dataset("penguins")
var
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var)
plt.show()
```

```python
# jaisa ki hum csv file me dekh sakte hai ki hamara data sex ke throw differentiate ho raha hai
# to hum yaha per pata ker sakte hai ki yaha per konsa deta male ka hai or konsa data female ka hai.
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex")
plt.show()
```

```python
#changing the style
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex",style="sex")
plt.show()
```

```python
#changing the size
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex",style="sex",size="sex")
plt.show()
```

```python
#changing the size according to him.
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex",style="sex",size="sex",sizes=(70,40))
plt.show()
```

```python
#changing the color
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex",style="sex",palette='gist_ncar')
plt.show()
```

```python
#alpha parameter
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex",style="sex",palette='gist_ncar',alpha=.5)
plt.show()
```

```python
#changing the marker
m={"Male":"*","Female":"o"}
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex",style="sex",palette='gist_ncar',markers=m)
plt.show()
```
