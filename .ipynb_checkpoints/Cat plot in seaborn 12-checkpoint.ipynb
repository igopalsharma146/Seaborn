# Cat plot in seaborn

```python
# catplot ek esa function hai jiske throw multiple plot ko ek single function ke throgh banaya ja sakta hai.
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# categorical scatterplots with catplot
    # stripplot()-with kind="strip"
    # swarmplot()-with kind="swarm"

# categorical estimate plots with catplot
    # pointplot()-with kind="point"
    # barplot()-with kind="bar"
    # countplot()-with kind="count"

# categorical distribution plots with cat plot
    # boxplot()-with kind="box"
    # violinplot()-with kind="violin"
    # boxenplot()-with kind="boxen"
```

```python
var=sns.load_dataset("tips").head(20)
var
```

```python
sns.catplot(x="tip",y="size",data=var)
plt.show()
```

```python
sns.catplot(x="tip",y="size",data=var,hue="sex")
plt.show()
```

```python
sns.catplot(x="tip",y="size",data=var,hue="sex",palette='Oranges')
plt.show()
```

```python
sns.catplot(x="tip",y="size",data=var,hue="sex",palette='Oranges',height=3)
plt.show()
```

```python
sns.catplot(x="tip",y="size",data=var,hue="sex",palette='Oranges',kind="bar")
plt.show()
```

```python
sns.catplot(x="day",y="size",data=var,hue="sex",palette='Oranges',kind="bar")
plt.show()
```

```python
sns.catplot(x="tip",y="size",data=var,hue="sex",palette='Oranges',kind="point") #box
plt.show()
```

```python
sns.catplot(x="day",y="size",data=var,hue="sex",palette='Oranges',kind="box") #box
plt.show()
```

```python
sns.catplot(x="day",y="size",data=var,hue="sex",palette='Oranges',kind="boxen") #box
plt.show()
```
