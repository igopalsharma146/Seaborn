# violin plot in seaborn

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.violinplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="time")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,linewidth=5)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,linewidth=3,palette='Dark2')
plt.show()
```

```python
#order parameter 
# order parameter ke throght hum graph ko different different order me covert kar sakte hai.
sns.violinplot(x="time",y="total_bill",data=var,order=["Dinner","Lunch"])
plt.show()
```

```python
#saturation parameter
sns.violinplot(x="day",y="total_bill",data=var,linewidth=3,palette='Dark2',saturation=0.6)
plt.show()
```

```python
# color parameter
sns.violinplot(x="day",y="total_bill",data=var,linewidth=3,color="y")
plt.show()
```

```python
#hue parameter
sns.violinplot(x="day",y="total_bill",data=var,linewidth=3,hue="sex")
plt.show()
```

```python
# split parameter 
sns.violinplot(x="day",y="total_bill",data=var,hue="sex",split=True)
plt.show()
```

```python
#scale parameter
sns.violinplot(x="day",y="total_bill",data=var,hue="sex",split=True,scale="count") 
# hum scale ke ander ("area","count","width") in sabko de sakte hai.
plt.show()
```

```python
#scale parameter
sns.violinplot(x="day",y="total_bill",data=var,hue="sex",split=True,scale="width") 
plt.show()
```

```python
#horizontal graph
sns.violinplot(x="total_bill",y="day",data=var,hue="sex") 
plt.show()
```

```python
#inner parameter
sns.violinplot(x="time",y="total_bill",data=var,order=["Dinner","Lunch"],inner="quart")
# inner parameter ke ander hum ("box","quartile","point","stick",None) etc. ka use kar sakte hai.
plt.show()
```

```python
# hoe to make a single violin plot
sns.violinplot(y="total_bill",data=var) 
plt.show()
```
