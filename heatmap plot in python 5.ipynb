# heatmap plot in python

```python
# jab hamre pass large amount of data hai, to hum heatmap ka use karke use color bar me show kara sakte hai.
#heatmap me hum jo bhi data dete hai, to us data ko hum color ke throw represent karte hai.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# var=sns.load_dataset("penguins")
# var
```

```python
var=np.linspace(1,10,20)
var
```

```python
var1=np.linspace(1,10,20).reshape(4,5)
var1
```

```python
sns.heatmap(var1)
```

```python
var2=sns.load_dataset("anagrams")
var2
```

```python
x=var2.drop(columns="attnr",axis=1).head(10)
x
```

```python
sns.heatmap(x)
```

```python
# hum yaha per heatmap ki minimum and maximum value ko change kar sakte hai, uske liye hum vmin or vmax parameter ka use kar sakte hai.
# sns.heatmap(x,vmin=4,vmax=10)
# sns.heatmap(x,vmin=0,vmax=5)
sns.heatmap(x,vmin=0,vmax=15)
```

```python
# cmap paramter ke throw hum heatmap ke color ko bhi change kar sakte hai.
# sns.heatmap(x,vmin=0,vmax=10,cmap="A")     # This will throw an error but by using this we can get the cmap.
sns.heatmap(x,vmin=0,vmax=10,cmap="gist_heat")
```

```python
sns.heatmap(x,vmin=0,vmax=10,cmap='PuOr')
```

```python
#agar hum chate hai ki hame heatmap ki value show ho, that mean jiskkki jo value ho vo yaha per show ho.uske liye hum annot parameter ka use kar sakte hai.
sns.heatmap(x,vmin=0,vmax=10,cmap="gist_heat",annot=True)
```

```python
# agar hum chahte hai ki in value ki jagah, hamari khud ki value print ho to hum vo bhi yaha per kar sakte hai .
#uske liye hame ek list banani padegi or us list ke under hum iski value ko show kara skte hai.
var3=np.linspace(1,10,10).reshape(2,5)
print(var3)
sns.heatmap(var3,vmin=0,vmax=15,cmap="gist_heat",annot=True)
```

```python
ar=np.array([["a0","a1","a2","a3","a4"],["b0","b1","b2","b3","b4"]])
ar
```

```python
sns.heatmap(var3,vmin=0,vmax=15,cmap="gist_heat",annot=ar,fmt="s") #hame yaha per hamare data ka formate bhi dena hoga.
```

```python
# hum yaha per iske font or color ko bhi change kar sakte hai.
y={"fontsize":15,"color":"r"}
sns.heatmap(var3,vmin=0,vmax=15,cmap="PuOr",annot=ar,fmt="s",annot_kws=y)
```

```python
# hum yaha per box ke line ki width ko bhi badha sakte hai.uske liye hum yaha per line width paramter ka use karenge.
sns.heatmap(var3,vmin=0,vmax=15,cmap="gist_heat",annot=True,linewidth=5)
```

```python
#changin the line color
sns.heatmap(var3,vmin=0,vmax=15,cmap="gist_heat",annot=True,linewidth=5,linecolor="b")
```

```python
# removing the colorbar
sns.heatmap(var3,vmin=0,vmax=15,cmap="gist_heat",annot=True,linewidth=5,linecolor="b",cbar=False)
plt.show()
```

```python
# removing the xticklabels
sns.heatmap(var3,vmin=0,vmax=15,cmap="gist_heat",annot=True,linewidth=5,linecolor="b",cbar=False,xticklabels=False,yticklabels=False)
plt.show()
```

```python
# now , we are giving the labels throw the seaborn
# hum yaha per matplotlib ke throw bhi labels de sakte hai but hum yaha per seaborn ke throw labels denge.
v=sns.heatmap(var3,vmin=0,vmax=15,cmap="gist_heat",annot=True,linewidth=5,linecolor="b",cbar=False,xticklabels=False,yticklabels=False)
v.set(xlabel="python",ylabel="MITRC")

plt.show()
```

```python
#changing the fontsize and color of our labels
v=sns.heatmap(var3,vmin=0,vmax=15,cmap="gist_heat",annot=True,linewidth=5,linecolor="b",cbar=False,xticklabels=False,yticklabels=False)
v.set(xlabel="python",ylabel="MITRC")
sns.set(font_scale=2)
plt.show()
```
