# Factor plot in seaborn

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
var=sns.load_dataset("tips")
var
```

```python
#ERROR : module 'seaborn' has no attribute 'factorplot'
#factorplot function has been removed to 'catplot'
sns.catplot(x="size",y="tip",data=var)
plt.show()
```

```python
sns.catplot(x="size",y="tip",data=var,hue="sex")
plt.show()
```

```python
sns.catplot(x="size",y="tip",data=var,hue="sex",kind="box")
plt.show()
```
