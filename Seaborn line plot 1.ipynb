# Seaborn line plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
var1=[1,2,3,4,5,6,7]
var2=[5,6,3,7,3,7,3]
plt.plot(var1,var2)
plt.show()
```

```python
x=pd.DataFrame({"var1":var1,"var2":var2})

sns.lineplot(x="var1",y="var2",data=x)
plt.show()
```

```python
x=pd.DataFrame({"var1":var1,"var2":var2})
sns.lineplot(x=var1,y=var2)
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1)
```

```python
y_1=sns.load_dataset("penguins")

#if we want to make graph of most popular data
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex")
```

```python
y_1=sns.load_dataset("penguins")

#if we want to make graph of most popular data
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",size=50)
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",style="sex")
```

```python
#hum yaha per palette ki help se color ko bhi change kar sakte hai.
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",style="sex",palette='Accent')
```

```python
#marker
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",style="sex",palette='Accent',markers=["o",">"])
plt.show()
```

```python
y_1=sns.load_dataset("penguins").head(20)
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",style="sex",palette='Accent',markers=["o",">"])
plt.show()
```

```python
#hum yaha per dashes parameter ka use karke hamare dashes ko hata sakte hai.
y_1=sns.load_dataset("penguins").head(20)
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",style="sex",palette='Accent',markers=["o",">"],dashes=False,legend=True)
plt.show()
```

```python
#hum yaha per dashes parameter ka use karke hamare dashes ko hata sakte hai.
y_1=sns.load_dataset("penguins").head(20)
plt.grid()
plt.title("python")
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",style="sex",palette='Accent',markers=["o",">"],dashes=False,legend=True)
plt.show()
```
