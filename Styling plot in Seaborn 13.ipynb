# Styling plot in Seaborn

```python
# Seaborn figure style
# Removing Axes Spines
# Scale and Context

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
var=sns.load_dataset("tips")
var
```

# Seaborn figure style

```python
# Seaborn figure style
sns.boxplot(y=var["total_bill"])
plt.show()
```

```python
sns.barplot(x="day",y="total_bill",data=var)
plt.grid()
plt.show()
```

```python
# making the grid using the seaborn
sns.set_style("darkgrid") # whitegrid,darkgrid,ticks
sns.barplot(x="day",y="total_bill",data=var)
plt.grid()
plt.show()
```

# Removing Axes Spines

```python
sns.set_style("white") # whitegrid,darkgrid,ticks
sns.barplot(x="day",y="total_bill",data=var)

#removing Axes Spines
sns.despine()
plt.show()
```

# Scale and Context

```python
# changing the figure size
sns.set_style("white") # whitegrid,darkgrid,ticks
plt.figure(figsize=(12,3))
sns.barplot(x="day",y="total_bill",data=var)

#removing Axes Spines
sns.despine()
plt.show()
```

```python
# changing the figure size
sns.set_style("white") # whitegrid,darkgrid,ticks
plt.figure(figsize=(3,9)) #showing vertically
sns.barplot(x="day",y="total_bill",data=var)

#removing Axes Spines
sns.despine()
plt.show()
```

```python
# changing the font
sns.set_style("white") # whitegrid,darkgrid,ticks
sns.set_context("poster",font_scale=1) #paper
sns.barplot(x="day",y="total_bill",data=var)

#removing Axes Spines
sns.despine()
plt.show()
```

```python
# changing the font
sns.set_style("white") # whitegrid,darkgrid,ticks
sns.set_context("poster",font_scale=1) #paper
sns.barplot(x="day",y="total_bill",data=var,palette='cool')

#removing Axes Spines
sns.despine()
plt.show()
```
