# box plot in seaborn

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
var=sns.load_dataset("tips")
var
```

```python
# box plot banane se pahle hame grid banani padegi
sns.set(style="whitegrid")
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
# differentiating according to the lunch and dinner
sns.boxplot(x="day",y="total_bill",data=var,hue="time")
plt.show()
```

```python
# differentiating according to the sex
sns.boxplot(x="day",y="total_bill",data=var,hue="sex",saturation=150)
plt.show()
```

```python
# changing the color
sns.boxplot(x="day",y="total_bill",data=var,saturation=150,color='g')
plt.show()
```

```python
# changing the color
sns.boxplot(x="day",y="total_bill",data=var,hue="sex",saturation=150,palette='plasma')
plt.show()
```

```python
# changing the order
sns.boxplot(x="day",y="total_bill",data=var,saturation=150,palette='plasma',order=["Fri","Sun","Thur","Sat"])
plt.show()
```

```python
# hum yaha per mean ko bhi represent kar sakte hai, uske liye hame showmeans name ke parameter ka use karna hoga.
# hum yaha per meanprops parameter ka use karke mean ki property ko change kar sakte hai.
sns.boxplot(x="day",y="total_bill",data=var,saturation=150,showmeans=True,meanprops={"marker":"+","markeredgecolor":"y"})
plt.show()
```

```python
# changing the linewidth
sns.boxplot(x="day",y="total_bill",data=var,saturation=150,showmeans=True,meanprops={"marker":"+","markeredgecolor":"y"},
           linewidth=3)
plt.show()
```

```python
# hum orient parameter ke throw iske orientation ko bhi change kar ssakte hai
sns.boxplot(x="day",y="total_bill",data=var,saturation=150,showmeans=True,meanprops={"marker":"+","markeredgecolor":"y"},
           linewidth=3,orient="v") 
# if we want to convert this graph is in horizontal, to uske liye hame yaha per x or y Axis ko change karna padega.
# because y axis per hamesha numerical data hona chahiye
plt.show()
```

```python
# hum orient parameter ke throw iske orientation ko bhi change kar ssakte hai
sns.boxplot(x="total_bill",y="day",data=var,saturation=150,showmeans=True,meanprops={"marker":"+","markeredgecolor":"y"},
           linewidth=3,orient="h") 
# if we want to convert this graph is in horizontal, to uske liye hame yaha per x or y Axis ko change karna padega.
# because y axis per hamesha numerical data hona chahiye
plt.show()
```

```python
sns.boxplot(data=var,saturation=150,showmeans=True,meanprops={"marker":"+","markeredgecolor":"y"},
           linewidth=3,orient="h") 
plt.show()
```

```python
# making a single box
sns.boxplot(x=var["total_bill"])
plt.show()
```

```python
# making a single box
sns.boxplot(y=var["total_bill"])
plt.show()
```
