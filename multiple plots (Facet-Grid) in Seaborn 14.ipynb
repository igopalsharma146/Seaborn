```python
# multiple plots (Facet-Grid) in Seaborn
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
var=sns.load_dataset("tips")
var
```

```python
fg=sns.FacetGrid(var)
fg.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
fg=sns.FacetGrid(var,col="sex")
fg.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
fg=sns.FacetGrid(var,col="sex",hue="day")
fg.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
fg=sns.FacetGrid(var,col="sex",hue="day")

#adding legend
fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
# agar hum day ke according alag alag graph chahte hai to vo bhi hum yaha per kar sakte hai.
fg=sns.FacetGrid(var,col="day",hue="sex")

#adding legend
fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
# agar hum scatter plot ki jagah per agar kooi or graph chahte hai to scatter ki jagah us graph ka name likh denge
fg=sns.FacetGrid(var,col="day",hue="sex")

#adding legend
fg.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```

```python
# agar hum scatter plot ki jagah per agar kooi or graph chahte hai to scatter ki jagah us graph ka name likh denge
fg=sns.FacetGrid(var,col="sex",hue="day")

#adding legend
fg.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```

```python
# agar hum scatter plot ki jagah per agar kooi or graph chahte hai to scatter ki jagah us graph ka name likh denge
fg=sns.FacetGrid(var,col="sex",hue="day",palette='cool')

#adding legend
fg.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```

```python
# cahnging the edgecolor
fg=sns.FacetGrid(var,col="sex",hue="day",palette='cool')

#adding legend
fg.map(plt.bar,"total_bill","tip",edgecolor="black").add_legend()
plt.show()
```
